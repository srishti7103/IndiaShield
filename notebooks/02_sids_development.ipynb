# 02. SIDS Metric Development

This notebook covers the formulas and calculations behind the **Strategic Import Dependency Score (SIDS)**, quantifying India's risk profile for defense procurements.

```python
import pandas as pd
import numpy as np

def calculate_sids(ic, ssr, gss, dsc):
    # SIDS = (IC/100) * SSR * (1 - GSS) * (1/DSC) * 100
    return (ic / 100.0) * ssr * (1.0 - gss) * (1.0 / dsc) * 100.0

# Sample calculation for Russia (post-2022)
score = calculate_sids(45, 0.4, 0.2, 0.2)
print(f"Russia SIDS raw: {score}")
```
